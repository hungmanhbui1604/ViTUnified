```python
import glob
import os
from PIL import Image
from tqdm import tqdm
```

```python
base = "data/LivDet"
paths = sorted(
    glob.glob(os.path.join(base, "**", "*.png"), recursive=True) +
    glob.glob(os.path.join(base, "**", "*.bmp"), recursive=True)
)

len(paths)
```

```python
broken_paths = []
for p in tqdm(paths, desc="Checking images"):
    try:
        img = Image.open(p).convert("RGB")
    except Exception as e:
        broken_paths.append((p, str(e)))

print(f"Total broken images: {len(broken_paths)}")
```
